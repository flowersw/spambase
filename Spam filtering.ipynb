```python
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_filter = pd.read_csv('spambase/spambase.data')
```

```python
spam_filter.head()
```

```python
spam_filter = pd.read_csv('spambase/spambase.data', header = None)
```

```python
spam_filter.head()
```

```python
#The last column (57) says whether it's spam or not
input_data = spam_filter.iloc[:, :57]

output_data = spam_filter.iloc[:, 57:]
```

```python
input_data.head()
```

```python
output_data.head()
```

```python
input_training_data, input_testing_data, output_training_data, output_testing_data = train_test_split(input_data, output_data, train_size=0.6, test_size=0.4)
```

```python
input_testing_data.head()
```

```python
multinomial_naive_bayes = MultinomialNB()
```

```python
multinomial_naive_bayes.fit(input_training_data, output_training_data)
```

```python
multinomial_naive_bayes.predict(input_testing_data)
```

```python
multinomial_naive_bayes.score(input_training_data, output_training_data)
```

```python
multinomial_naive_bayes.score(input_testing_data, output_testing_data)
```

```python
#Pretty close: Both almost 80% correct
```
